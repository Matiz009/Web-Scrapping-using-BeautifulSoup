```python
import requests
from bs4 import BeautifulSoup
url = 'https://en.m.wikipedia.org/wiki/Le_Monde%27s_100_Books_of_the_Century'
page = requests.get(url)
soup = BeautifulSoup(page.content,'html.parser')
print(soup)  #getting the html page and printing it
```

```python
print(soup.prettify())  #removing unnecessary data
```

```python
tbody = soup.tbody
print(tbody)  #getting the data in tables
```

```python
table=soup.find('table',{'class':'headerSort'})
```
